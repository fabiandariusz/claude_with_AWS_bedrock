```python
import boto3
```

```python
import os
os.environ['AWS_BEARER_Token_BEDROCK'] = ""
```

```python
client = boto3.client("bedrock-runtime", region_name="us-east-2")
model_id = "us.anthropic.claude-3-5-haiku-20241022-v1:0"

def add_user_message(messages, text):
    user_message = {
        "role": "user",
        "content": [{"text": text}]
    }
    messages.append(user_message)


def add_assistant_message(messages, text):
    assistant_message = {
        "role": "assistant", 
        "content": [{"text": text}]
    }
    messages.append(assistant_message)

def chat(messages, system=None, temperature=0.3):
    params = {
        "modelId": model_id, 
        "messages": messages,
        "inferenceConfig": { "temperature": temperature}
    }

    if system:
        params["system"] = [{"text":system}]

    response = client.converse(**params)

    return response["output"]["message"]["content"][0]["text"]
```

```python
messages = []

add_user_message(messages, "write a python function that checks a string for duplicate charecters")
text = chat(messages, system="You are python developer who writes concise code")

print(text)
```
